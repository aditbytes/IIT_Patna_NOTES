## Hash Function

A hash function is a mathematical function that converts input data (called a key) into a fixed-size numerical value, usually called a hash value or hash code.
It’s widely used in computer science for data storage, cryptography, and efficient searching.

You can think of it as:

Input → [Hash Function] → Output (Hash value)

For example:

    h("apple") = 3
    h("banana") = 7
    h("cherry") = 2

### Why Do We Use Hash Functions?

Imagine you have a list of 1 million names and you want to quickly check if “Modi” is present.
Instead of checking each name one by one, you can hash the name and jump directly to its position in a table — this makes searching much faster (on average O(1) time).

    h(23) = 3
    h(43) = 3


Both keys go to index 3 — the hash function can’t distinguish between them.

## Collisions

When two different keys have the same hash value, it’s called a collision.
A good hash function minimizes collisions.

Common collision handling methods:

- Chaining – Store multiple elements in a linked list at the same index.

- Open addressing – Find another empty slot using a rule (like linear probing).

##  Collision resolution techniques

We can broadly divide hashing techniques into two categories:

- Open Hashing (Chaining)

- Closed Hashing (Open Addressing)

### Closed Hashing (Open Addressing)

Here, all elements are stored directly inside the hash table itself.
When a collision happens, the algorithm searches for the next available slot.

Types of Open Addressing:

- Linear Probing

- Quadratic Probing

- Double Hashing

## Open Hashing (Chaining Method)

Each index in the hash table stores a linked list (or a list) of keys that hash to the same index.

So if multiple keys have the same hash value, they are stored in the same bucket.

image.png

Example:

h = HashTable(5)
keys = [11, 12, 21, 22, 31]

Table Size = 5

Method:
Hashing with Chaining

Hash Function:
h(k) = k mod 5

Insert keys one by one:

1) Key = 11
h(11) = 11 mod 5 = 1
Place 11 in chain at index 1

2) Key = 12
h(12) = 12 mod 5 = 2
Place 12 in chain at index 2

3) Key = 21
h(21) = 21 mod 5 = 1
Collision at index 1
Add 21 to chain at index 1

4) Key = 22
h(22) = 22 mod 5 = 2
Collision at index 2
Add 22 to chain at index 2

5) Key = 31
h(31) = 31 mod 5 = 1
Collision at index 1
Add 31 to chain at index 1

Final Hash Table (with chains):

Index 0 : Empty
Index 1 :  11-> 21 -> 31
Index 2 : 12 -> 22
Index 3 : Empty
Index 4 : Empty

```python
def simple_hash(key, table_size):
    return key % table_size #=>k % m

keys = [17, 56, 21, 45, 48]
table_size = 10
# 17=>17%10=7

for k in keys:
    print(f"Key {k} → Hash value: {simple_hash(k, table_size)}")
```

```python
def folding_hash(key):
    key_str = str(key)  # convert number to string====> 123=> "123"
    total = 0

    # split key into 2-digit parts and add them
    for i in range(0, len(key_str), 2):
        part = int(key_str[i:i+2])# "1234"=> 12  34
        total += part  #total=0 + 12 + 34 =

    # take last two digits as hash value
    hash_value = total % 100
    return hash_value

# Example
keys = [123456, 987654, 654321, 111222]

for k in keys:
    print(f"Key {k} → Hash value: {folding_hash(k)}")
```

```python
# Mid-Square Method (Simple Version)

def mid_square_hash(key):
    square = key * key          # step 1: square the key
    square_str = str(square)     # step 2: convert to string 12^2=144===> "144"
    mid = len(square_str) // 2  # find middle position

    # step 3: take 2 middle digits
    middle_digits = square_str[mid-1:mid+1]  "14"===>14
    hash_value = int(middle_digits)
    return hash_value % 100     # step 4: modulo for table size

# Example
keys = [12, 23, 45, 67, 89]

for k in keys:
    print(f"Key {k} → Hash value: {mid_square_hash(k)}")
```

```python
def string_hash(s, table_size):
    total = 0
    for ch in s:
        total += ord(ch)   # convert char to ASCII
    return total % table_size

words = ["apple", "banana", "cherry"]
table_size = 10

for w in words:
    print(f"{w} → Hash value: {string_hash(w, table_size)}")
```

```python
# Simple Hash Table using Linear Probing

# create an empty hash table
size = 10
hash_table = [None] * size

# define hash function
def hash_function(key):
    return key % size

# insert function using linear probing
def insert(key):
    index = hash_function(key)

    # if slot is occupied, move to next
    while hash_table[index] is not None:
        index = (index + 1) % size  # go to next index (circular)
    hash_table[index] = key

# display hash table
def display():
    for i in range(size):
        print(i, ":", hash_table[i])

# example keys
keys = [23, 43, 13, 27]

for k in keys:
    insert(k)

display()
```

```python
# Quadratic Probing
class QuadraticProbingHash:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return key % self.size

    def insert(self, key):
        index = self.hash_function(key)
        i = 0
        while self.table[(index + i * i) % self.size] is not None:
            i += 1
        self.table[(index + i * i) % self.size] = key

    def display(self):
        for i, v in enumerate(self.table):
            print(i, ":", v)

# Example
h = QuadraticProbingHash(10)
for k in [23, 43, 13, 27]:
    h.insert(k)
h.display()
```

```python
# Double Hashing
class DoubleHashing:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def h1(self, key):
        return key % self.size

    def h2(self, key):
        return 7 - (key % 7)   # second hash (prime less than size)

    def insert(self, key):
        index = self.h1(key)
        step = self.h2(key)
        i = 0
        while self.table[(index + i * step) % self.size] is not None:
            i += 1
        self.table[(index + i * step) % self.size] = key

    def display(self):
        for i, v in enumerate(self.table):
            print(i, ":", v)

# Example
h = DoubleHashing(10)
for k in [23, 43, 13, 27]:
    h.insert(k)
h.display()
```

```python
# Hashing with Chaining
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):
        return key % self.size

    def insert(self, key):
        index = self.hash_function(key)
        self.table[index].append(key)

    def display(self):
        for i in range(self.size):
            print(i, ":", self.table[i])

# Example
h = HashTable(5)
keys = [11, 12, 21, 22, 31]
for k in keys:
    h.insert(k)
h.display()
```
