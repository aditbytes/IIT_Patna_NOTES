```python
# STACK using class

class Stack:
  def __init__(self):
    self.stack=[]

  def push(self,element):
    self.stack.append(element)

  def isEmpty(self):
    return len(self.stack)==0

  def pop(self):
    if self.isEmpty():
      return "Stack is empty"
    return self.stack.pop()

  def peek(self):
    if self.isEmpty():
      return "Stack is Empty"
    return self.stack[-1]



#  Create a Stack

P_stack= Stack()
P_stack.push("A")
P_stack.push("B")
P_stack.push("C")

print("STACK:", P_stack.stack)
print("POP:", P_stack.pop())
print("isEMPTY:", P_stack.isEmpty())
print("PeeK:", P_stack.peek())



```

```python
# Doubly Linked List using class


class Node_D:
  def __init__(self,data):
    self.data= data
    self.prev= None    # One addition in Singly linked list
    self.next= None    # For next node pointer


class DoubleLinkedList:
  def __init__(self):
    self.head=None      # Head Ponts to the first node

  # Defnition for forward Traversal

  def display_forward(self):
    temp= self.head

    while temp:
      print(temp.data, end="<-->")
      temp=temp.next
    print("None")


  # Defnition to Insert at the beginning of a Doubly Linked List

  def insert_at_beginning(self,data):
    new_node= Node_D(data)

    if self.head is None:
      self.head=new_node
      return

    new_node.next=self.head
    self.head.prev=new_node
    self.head=new_node

  # Definition for Insert at the end

  def insert_at_end(self,data):
    new_node=Node_D(data)

    if self.head is None:
      self.head=new_node

    temp=self.head
    while temp.next:
      temp=temp.next

    temp.next=new_node
    new_node.prev=temp

  # Defnition to add inside DLL

  def insert_at_position(self, position, data):
    new_node= Node_D(data)

    if position==0:                 #Position==0 means insert at the beginning
      if self.head:
        self.head.prev=new_node
        new_node.next=self.head
      self.head=new_node
      return

    temp=self.head
    count=0

    while temp and count < position-1:
      temp=temp.next
      count +=1

    if temp is None:
      print("Position out of range")
      return

    new_node.next=temp.next
    new_node.prev=temp

    #if temp.next:
     # temp.next.prev=new_node

    temp.next=new_node

  # Defnition to delete node at given position

  def delete_at_position(self, position):
    if self.head is None:
      print("DLL emplty")
      return

    if position==0:
      self.head=self.head.next
      if self.head:
        self.head.prev=None
      return

    temp=self.head
    count=0

    while temp and count< position:
      temp=temp.next
      count+=1

    if temp is None:
      print("Position out of range")

    if temp.prev:
      temp.prev.next=temp.next
    if temp.next:
      temp.next.prev=temp.prev









# Creation of Linked List

dll=DoubleLinkedList()

dll.insert_at_beginning(10)
dll.insert_at_beginning(20)
dll.insert_at_beginning(30)
dll.insert_at_end(50)
dll.display_forward()
dll.insert_at_end(100)
dll.display_forward()

dll.insert_at_position(3,200)
dll.display_forward()

dll.delete_at_position(0)
dll.display_forward()



```

```python
# Binary Search Tree

class BT_Node:
  def __init__(self, data):
    self.data=data
    self.left=None
    self.right=None

class BST:
  def __init__(self):
    self.root=None

  def insert(self,data):
    if self.root is None:
      self.root=BT_Node(data)
    else:self._insert(self.root, data)

  def _insert(self, current, data):
    if data < current.data:
      if current.left is None:
        current.left=BT_Node(data)
      else:
        self._insert(current.left, data)

    elif data > current.data:
      if current.right is None:
        current.right= BT_Node(data)
      else:
        self._insert(current.right, data)

#   Defnition for Minimum value in BST
  def find_min(self,node):
    current=node
    while current.left:
      current=current.left
    return current

# Defnition for Deletion of a Node in BST

  def delete(self, root, key):

    if root is None:
        return root

    if key < root.data:
        root.left = self.delete(root.left, key)

    elif key > root.data:
        root.right = self.delete(root.right, key)

    else:
        # Case 1 & 2
        if root.left is None:
            return root.right

        elif root.right is None:
            return root.left

        # Case 3 (two children)
        successor = self.find_min(root.right)
        root.data = successor.data
        root.right = self.delete(root.right, successor.data)

    return root









  def inorder(self, root):
    if root:
      self.inorder(root.left)
      print(root.data, end=" ")
      self.inorder(root.right)



##################

B_tree=BST()

B_tree.insert(10)
B_tree.insert(5)
B_tree.insert(20)
B_tree.insert(3)
B_tree.insert(7)
B_tree.insert(13)
B_tree.insert(14)
B_tree.insert(8)

B_tree.inorder(B_tree.root)

print("\n After deletion:")

B_tree.root=B_tree.delete(B_tree.root, 3)

B_tree.inorder(B_tree.root)



```

```python
from os import XATTR_CREATE
#  AVL Tree Implementation

# Node Class (with height)
# Insert Function
# Rotations (LL, RR, LR, RL)
# Inorder Taversal

class Node_A:
  def __init__(self, data):
    self.data= data
    self.left=None
    self.right=None
    self.height=1  # important for AVL

class AVLTree:

  def get_height(self,node):
    if not node:
      return 0
    return node.height

  # Code for Balance Factor
  def get_balance(self, node):
    if not node:
      return 0
    return self.get_height(node.left)- self.get_height(node.right)

  # Code for LL case (Right Rotation)

  def right_rotate(self,y):
     x =y.left                 #  y is a unbalanced node
     T2 = x.right
     # rotate
     x.right=y
     y.left=T2

     # Update height
     y.height= 1+ max(self.get_height(y.left), self.get_height(y.right))
     x.height=1+max(self.get_height(x.left), self.get_height(x.right))

     return x

  # Code for RR case (Left Rotation)
  def left_rotate(self, x):
    y = x.right
    T2 = y.left

    y.left=x
    x.right=T2

    x.height= 1+ max(self.get_height(x.left), self.get_height(x.right))
    y.height=1+max(self.get_height(y.left), self.get_height(y.right))

    return y

  # Code for insertion

  def insert(self, root, key):
    if not root:
      return Node_A(key)

    if key < root.data:
      root.left=self.insert(root.left, key)
    elif key > root.data:
      root.right=self.insert(root.right, key)
    else:
      return root

    root.height=1+max(self.get_height(root.left), self.get_height(root.right))
    balance=self.get_balance(root)

    # LL
    if balance > 1 and key < root.left.data:
      return self.right_rotate(root)

    # RR
    if balance < -1 and key > root.right.data:
      return self.left_rotate(root)

    # LR
    if balance > 1 and key > root.left.data:
      root.left=self.left_rotate(root.left)
      return self.right_rotate(root)

    # RL

    if balance < -1 and key <root.right.data:
      root.right= self.right_rotate(root.right)
      return self.left_rotate(root)

    return root







# Defnition for traversal
  def inorder(self,root):
    if root:
      self.inorder(root.left)
      print(root.data, end=" --")
      self.inorder(root.right)

#  Example run

avl=AVLTree()
root=None
nums=[10, 20, 30, 5, 15, 7, 21, 6, 1, 25, 2]
for num in nums:
  root=avl.insert(root, num)

print("Inorder Traversal:")
avl.inorder(root)












```

```python
# Graph Using Adjacency List

from collections import deque


# Step 1: Create a graph

graph={
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'E'],
    'D': ['B'],
    'E': ['C']
}


# Step 2: BFS (Breadth First Search)
def BFS(start):
  visited=[]
  queue=[] # Using list

  visited.append(start)
  queue.append(start)

  print("BFS Traversal")
  while queue:
    node=queue.pop(0) # remove first element
    print(node, end="->")

    for neighbour in graph[node]:  # go throufg all neighbours of node
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)

  # Def for DFS--------------------

def DFS(node, visited):
  if node not in visited:
    print(node,end="->")
    visited.append(node)      # Mark Visited

    for neighbour in graph[node]:
      DFS(neighbour, visited)  # recursive call

def DFS_Stack(node): # DFS using Stack
  visited=[]          # tack visited nodes
  stack=[]            # Stack (LIFO)

  stack.append(node)   # push start node
  print("DFS using Stack:")

  while stack:                       # Loop until stack empty
    node=stack.pop()                 # remove last element

    if node not in visited:         # if not visited
      print(node, end="->")         # print node
      visited.append(node)          # mark visited
      for neighbour in graph[node]:
        stack.append(neighbour)




#---------Main Program-----

print("Graph")
for node in graph:
  print(node, "->", graph[node])

print()

BFS('A')

print("\n DFS Traversal")
DFS('A', [])

DFS_Stack('A')
```

```python
# BFS and DFS on Graph using Class

class Graph:
  def __init__(self):
    self.graph={}         # dictionary to store graph

  def add_edge(self, u, v):
    if u not in self.graph:
      self.graph[u]=[]     # create empty list

    if v not in self.graph:
      self.graph[v]=[]     # create empty list


    self.graph[u].append(v)    # add v to u list
    self.graph[v].append(u)    # add u in v list

  # Step 2: BFS (Breadth First Search)
  def bfs(self, start):
    visited=[]
    queue=[] # Using list

    visited.append(start)
    queue.append(start)

    print("BFS Traversal")
    while queue:
      node=queue.pop(0) # remove first element
      print(node, end="->")

      for neighbour in self.graph[node]:  # go throufg all neighbours of node
        if neighbour not in visited:
          visited.append(neighbour)
          queue.append(neighbour)

#--------------  DFS---------
  def dfs(self, node, visited):
    if node not in visited:
      print(node,end="->")
      visited.append(node)      # Mark Visited

      for neighbour in self.graph[node]:
        self.dfs(neighbour, visited)  # recursive call


# ----------Main Programme------

g=Graph()           # create object
g.add_edge('A','B')
g.add_edge('A','C')
g.add_edge('B','D')

print("GRAPH")
for node in g.graph:
  print(node, "->", g.graph[node])
print()

g.bfs('A')

print("\n DFS Traversal")
g.dfs('A', [])
```

```python
#--- Selection Sort----

def selection_sort(arr):
  n=len(arr)

  for i in range(n):
    min_index=i

    for j in range(i+1,n):
      if arr[j] < arr[min_index]:
        min_index=j

    # Swap
    arr[i], arr[min_index] = arr[min_index], arr[i]
    #

  return arr

#-----Main Programme---

A=[7, 12,9,11,3]
print(selection_sort(A))


```

```python
# -----------------Insertion Sort---------------

def insertion_sort(arr):
  for i in range (1, len(arr)):  # loop from 2nd element to last
    j=i

    while j>0 and arr[j-1] > arr[j]: # while left element is bigger
      temp=arr[j]
      arr[j]=arr[j-1]
      arr[j-1]=temp
      j=j-1          #move one step left
  return arr


#-----------Main Programme------------

A=[4,7,2,9,5,7,1,8]
print(insertion_sort(A))

```

```python
# Quick Sort Algorithm

def quick_sort(arr):
  if len(arr)<=1:
    return arr

  pivot=arr[0]
  rest=arr[1:]
  left=[]
  right=[]

  for i in rest:
    if i<=pivot:
      left.append(i)
    else:
      right.append(i)

  left_sorted=quick_sort(left)
  right_sorted=quick_sort(right)

  result= left_sorted+[pivot]+right_sorted
  return result

#---------Main Programme------
A=[4,6,8,2,9,1,8,0]
print(quick_sort(A))
```

```python
#--------Linear Search------------

def linear_search(arr, target):
  for i in range(len(arr)):     #  loop through all indices of the list
    if arr[i]==target:
      return i

  return -1                   # if loop ends, target not found

#--------Main Programme--------

A=[4, 2,7,1,9,8,3]
target=5

result=linear_search(A,target)
print(result)
```

```python
# ------Dijikstra Algorithm----
graph={
    'A':[('B',1), ('C',4)],
    'B':[('A',1), ('C',2), ('D',5)],
    'C':[('A',4), ('B',2)],
    'D':[('B',5), ('E',3)],
    'E':[]
}



def dijkstra_simple(graph, start):   # Function to find shortest path from start node
  distances={}                     # Create a dictionary to store shortest distances

  # Step 1:  Initialize all distances to infinity

  for node in graph:
    distances[node]=float('inf')     # Set initial distances as infinity

  distances[start]=0                 # distance of starting node is 0

  visited=[]                # list to keep track of visited node

  # loop untill all nodes are visited

  while len(visited) < len(graph): # continue untill all nodes are processed

  # Step 2: find unvisited node with minimum distance

    min_node=None     # Variable to store noe with smallest distance
    min_distance=float('inf')  # start with infinity as minimum

    for node in graph:  # check each node in graph
      if node is not visited and distances[node] < min_distance:
        min_distance=distances[node]
        min_node=node

    # Step-3: mark this node as visited
    visited.append(min_node)

    # Step 4: Update neighbours of selected node

    for neighbour, weight in graph[min_node]:
      new_distance=distances[min_node]+weight
      if new_distance < distances[neighbour]:
        distances[neighbour]=new_distance

  return distances



#------Main progamme----
result=dijkstra_simple(graph, 'A')
print(result)
```

```python
from ast import MatchStar
# Prim's Algorithm------------

def prims(graph, start):
  visited=[start]        # start node is marked visited

  mst=[]

  while len(visited) < len(graph): #run untill all nodes covers
    min_edge=None    # store smallest edge in neighbour
                     # it is a tuple and values are (node, neighbour, weight)

    # check all edges from visited nodes
    for node in visited:
      for neighbor in graph[node]:
        if neighbor not in visited:
          weight=graph[node][neighbor]

          # choose smallest edge
          if min_edge is None or weight < min_edge[2]:
            min_edge=(node, neighbor, weight)

    mst.append(min_edge)
    visited.append(min_edge[1])

  return mst

#--------------------------

graph={
    'A':{'B':2, 'C':3},
    'B':{'A':2, 'C':1, 'D':4},
    'C':{'A':3, 'B':1, 'D':5},
    'D':{'B':4, 'C':5},
}

result=prims(graph, 'A')
print(result)


```

```python
# Hash Table

class HashTable:
  def __init__(self,size):
    self.size=size
    self.table=[None]*size

  def hash_function(self,key):
    return key%self.size

  def insert(self,key):
    index=self.hash_function(key)
    if self.table[index] is None:
      self.table[index]=[key]
    else:
      self.table[index].append(key)

  def search(self,key):
    index=self.hash_function(key)
    if self.table[index] is None:
      return False
    else:
      return key in self.table[index]


  def display(self):
    for i in range(self.size):
      print(i, "->", self.table[i])


#..........Main Programm

ht=HashTable(3)

ht.insert(23)
ht.insert(3)
ht.insert(23)
ht.insert(26)
ht.insert(43)
ht.insert(17)
ht.insert(14)


#print(ht.search (53))
ht.display()
```

```python
# Hash Set

class HashSet:
  def __init__(self, size):
    self.size=size
    self.table=[[]for _ in range(size)]   # Bucket

  def hash_function(self,key):
    return key%self.size

  def add(self,key):
    index=self.hash_function(key)
    if key not in self.table[index]:
      self.table[index].append(key)


  def remove(self, key):
    index=self.hash_function(key)
    if key in self.table[index]:
      self.table[index].remove(key)

  def contains(self, key):
    index=self.hash_function(key)
    return key in self.table[index]

  def display(self):
    for i, bucket in enumerate(self.table):
      print(i, "->", bucket)

#-----------------Main Programme---

hs=HashSet(10)

hs.add(10)
hs.add(40)
hs.add(20)
hs.add(10)
hs.add(30)

hs.display()

```
