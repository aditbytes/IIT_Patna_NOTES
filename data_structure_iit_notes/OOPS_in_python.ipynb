## OOPS

Object-Oriented Programming is a programming paradigm where we model programs using objects that represent real-world entities.

Instead of only thinking in terms of functions and data separately, OOPS combines data (attributes) and behavior (methods) into a single unit called a class.



Object = (State, Behavior)

Where:

State → variables

Behavior → functions

## Class and Object

### Class

A class is a blueprint or template.

### Object

An object is an instance of a class.



Example analogy:

Class → Car

Object → Ford, Toyata

```python
# Class definition
class Car:
    def __init__(self, brand, color):
        # Properties (attributes)
        self.brand = brand
        self.color = color
        self.speed = 0

    # Methods (behavior)
    def start(self):
        print(f"{self.brand} car has started")

    def accelerate(self, value):
        self.speed += value
        print(f"{self.brand} speed is now {self.speed} km/h")
```

```python
# Objects of Car class
ford = Car("Ford", "Blue")
toyota = Car("Toyota", "White")

bmw=Car("BMW", "White")
```

```python
ford.start() ##using object we will call function
ford.accelerate(30)

toyota.start()
toyota.accelerate(50)

bmw.start()
```

## Encapsulation


Encapsulation is about protecting data inside a class.

It means keeping data (properties) and methods together in a class, while controlling how the data can be accessed from outside the class.

This prevents accidental changes to your data and hides the internal details of how your class works.

image.png

A class is an example of encapsulation as it encapsulates all the data that is member functions, variables, etc.



n Python, this is done using naming conventions:

- _var → protected

- __var → private

```python
class BankAccount:
    def __init__(self, account_holder, initial_balance):
        self.account_holder = account_holder      # public
        self.__balance = initial_balance           # private (encapsulated)

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive")
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive")
        if amount > self.__balance:
            raise ValueError("Insufficient balance")
        self.__balance -= amount

    def get_balance(self):
        return self.__balance
```

```python
acc= BankAccount("Rahul", 1000)
acc.deposit(500)
acc.withdraw(300)

#acc.__balance=1000000
print(acc.get_balance())
```

## Polymorphism

Polymorphism in Python means "same operation, different behavior." It allows functions or methods with the same name to work differently depending on the type of object they are acting upon.


image.png

```python
class Shape:
    def calculateArea(self):
        pass
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculateArea(self):# same method
        return 3.14 * self.radius * self.radius #behaviour diff( Circle area)


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculateArea(self):
        return self.length * self.width


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def calculateArea(self):
        return 0.5 * self.base * self.height


shapes = [
    Circle(5),
    Rectangle(10, 4),
    Triangle(6, 8)
]

for shape in shapes:
    print(shape.calculateArea())
```

## Encapsulation code

A company wants to manage details of newly joined employees.

Each employee has:

- employee_id (int)

- name (string)

- basic_salary (float)

- department (string)

The company follows these rules:

1. HRA = 20% of basic salary

2. DA = 10% of basic salary

3. Gross Salary = basic + HRA + DA

  If gross salary > 80,000 → employee is Senior Level Else → Junior Level

###Tasks:

- Create a class Employee

- Use a constructor to initialize employee details

- Create methods to:

   - calculate gross salary

    - determine employee level

    - display employee details

```python
class Employee:
    def __init__(self, emp_id, name, basic_salary, department):
        self.emp_id = emp_id
        self.name = name
        self.basic_salary = basic_salary
        self.department = department

    def calculate_gross_salary(self):
        hra = 0.20 * self.basic_salary
        da = 0.10 * self.basic_salary
        gross_salary = self.basic_salary + hra + da
        return gross_salary

    def employee_level(self):
        gross_salary = self.calculate_gross_salary()
        if gross_salary > 80000:
            return "Senior Level"
        else:
            return "Junior Level"

    def display(self):
        print("Employee ID:", self.emp_id)
        print("Name:", self.name)
        print("Department:", self.department)
        print("Basic Salary:", self.basic_salary)
        print("Gross Salary:", self.calculate_gross_salary())
        print("Level:", self.employee_level())
```

```python
emp1= Employee(101, "Amit", 65000, "IT")
emp2= Employee(102, "Aditya", 90000, "Electronics")
```

```python
emp1.display()
```

```python
emp2.display()
```

## Inheritance code

A company manages devices used by employees.

Every device has:

- brand

- model

There are different types of devices.

A Laptop is a type of Device and has an extra feature:

- battery_life (in hours)

A MobilePhone is also a type of Device and has an extra feature:

- camera_quality (in MP)

Tasks

- Create a base class Device

- Create two derived classes:

    - Laptop

    - MobilePhone

Use inheritance so that:

- Common properties are in Device

- Specific properties are in child classes

- Create objects of Laptop and MobilePhone

- Display their details

```python
class Device:
  def __init__(self, brand, model):
    self.brand=brand
    self.model=model

  def show_device_info(self):
    print("Brand:", self.brand)
    print("Model:", self.model)

class laptop(Device):
  def __init__(self, brand, model, battery_life):
    super().__init__(brand, model)
    self.battery_life=battery_life

  def show_laptop_info(self):
    self.show_device_info()
    print("Battery Life:", self.battery_life, "hours")

class MobilePhone(Device):
  def __init__(self, brand, model, camera_quality):
    super().__init__(brand, model)
    self.camera_quality=camera_quality

  def show_mobile_phone_info(self):
    self.show_device_info()
    print("Camera Quality:", self.camera_quality, "MP")
```

```python
laptop1 = laptop("DELL", "Inspiron345", 10)
```

```python
laptop1.show_laptop_info()
```

```python
mobile1= MobilePhone("Samsung", "Galaxy", 50)
```

```python
mobile1.show_mobile_phone_info()
```

### Abstraction code

```python
# Import ABC (Abstract Base Class) and abstractmethod
from abc import ABC, abstractmethod


# Abstract Base Class
class Vehicle(ABC):
    def __init__(self, make, model, year):
        # Initializing common attributes
        self.make = make
        self.model = model
        self.year = year

    # Abstract method (no implementation here)
    @abstractmethod
    def display_info(self):
        pass
        # This method MUST be implemented by child classes


# Child class inheriting Vehicle
class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        # Call parent class constructor to initialize make, model, year
        super().__init__(make, model, year)

        # Initialize Car-specific attribute
        self.num_doors = num_doors

    # Implementing abstract method
    def display_info(self):
        print(f"Car: {self.year}, {self.make}, {self.model}, {self.num_doors} doors")
```

```python
# Creating object of child class (allowed)
car = Car("Porsche", "Cayenne S", 2022, 4)

# Calling implemented abstract method
car.display_info()
```

A shopping app supports different payment methods.

Every payment method must implement a method called pay().

- CreditCardPayment → prints “Payment done using Credit Card”

- UpiPayment → prints “Payment done using UPI”

###Tasks

- Create an abstract class Payment

- Add an abstract method pay()

- Create child classes CreditCardPayment and UpiPayment

- Implement pay() in both child classes

```python
# Import required modules for abstraction
from abc import ABC, abstractmethod


# Abstract base class
class Payment(ABC):

    # Abstract method (no implementation here)
    @abstractmethod
    def pay(self):
        pass
        # This method must be implemented by all child classes


# Child class 1
class CreditCardPayment(Payment):

    # Implementing abstract method
    def pay(self):
        print("Payment done using Credit Card")


# Child class 2
class UpiPayment(Payment):

    # Implementing abstract method
    def pay(self):
        print("Payment done using UPI")
```

```python
# Creating objects of child classes
payment1 = CreditCardPayment()
payment2 = UpiPayment()

# Calling pay method
payment1.pay()
payment2.pay()
```

A Zoo has different animals.

Every animal can make a sound, but the sound is different for each animal.

- Dog → barks

- Cat → meows

- Cow → moos

All animals have a method called sound().

###Task

1. Create a base class Animal

2. Define a method sound() in the base class

3. Create child classes:

    - Dog

    - Cat

    - Cow

4. Override the sound() method in each child class

5. Use a loop to call sound() for different animals

```python
# Base class
class Animal:
    def sound(self):
        print("Animal makes a sound")


# Child class 1
class Dog(Animal):
    def sound(self):
        print("Dog barks")


# Child class 2
class Cat(Animal):
    def sound(self):
        print("Cat meows")


# Child class 3
class Cow(Animal):
    def sound(self):
        print("Cow moos")
```

```python
# Creating objects of different animals
animals = [Dog(), Cat(), Cow()]

# Calling sound() using loop
for animal in animals:
    animal.sound()
```

## 🏦 Mini Project: Bank Management System

Problem Statement: Design a simple Bank System where:

- Every account has balance

- Balance should NOT be accessed directly

- Different accounts behave differently

- Common operations should be shared

image.png

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance   # private variable

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount.     #balance= balance + amount
            print(f"Deposited ₹{amount}")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        # Default behavior
        if amount <= self.__balance:
            self.__balance -= amount.  #balance=balance-amount
            print(f"Withdrawn ₹{amount}")
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.__balance
```

```python
#Savings account has withdrawal limit: Maximum ₹10,000 withdrawal at once

class SavingsAccount(BankAccount):
    def withdraw(self, amount):
        if amount > 10000:
            print("Savings account withdrawal limit exceeded")
        else:
            super().withdraw(amount)
```

```python
#Current accounts allow overdraft: Can withdraw extra ₹5,000 even if balance is low

class CurrentAccount(BankAccount):
    def withdraw(self, amount):
        if amount <= self.get_balance() + 5000:
            print(f"Withdrawn ₹{amount} (Overdraft allowed)")
        else:
            print("Overdraft limit exceeded")
```

```python
savings = SavingsAccount("Amit", 20000)
current = CurrentAccount("Ravi", 5000)

savings.withdraw(12000)   # limit exceeded
savings.withdraw(8000)

current.withdraw(9000)    # overdraft allowed
current.withdraw(12000)   # overdraft exceeded

print("Savings Balance:", savings.get_balance())
print("Current Balance:", current.get_balance())
```
