```python
# Reverse a string using Stack:

def reverse_s(string):
    stack=[]

    for char in string:
        stack.append(char)

    reversed_string=''
    while stack:
        reversed_string+=stack.pop()

    return reversed_string

String=input("Enter a string")
result=reverse_s(String)
print("Reversed string:",result)


```

```python
# Balanced Paranthesis:

def is_balanced(expr):
    stack=[]

    bracket_map={')':'(',']':'[','}':'{'}

    for char in expr:
        if char in "({[":
            stack.append(char)

        elif char in ")}]":
            if not stack:
                return "Not Balanced"

            top=stack.pop()
            if bracket_map[char]!=top:
                return "Not Balanced"

    if not stack:
        return "Balanced"
    else:
        return "Not Balanced"

expr=input("Enter an expression")
result=is_balanced(expr)
print(result)
```

**Queue**:A Queue is a linear data structure that follows the principle of
FIFO (First In, First Out).

```python
# Ticket Counter Simulation using Queue (FIFO)

queue = []

def enqueue():
    name = input("Enter customer name: ")
    queue.append(name)
    print(name, "has joined the ticket line.")

def dequeue():
    if not queue:
        print("No customers in line.")
    else:
        served = queue.pop(0)
        print(served, "got the ticket and left the line.")

def display():
    if not queue:
        print("The queue is empty.")
    else:
        print("Customers in line:", queue)
while True:
    print("\n Ticket Counter ")
    print("1. Add Customer ")
    print("2. Serve Customer ")
    print("3. Display Queue")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        enqueue()
    elif choice == '2':
        dequeue()
    elif choice == '3':
        display()
    elif choice == '4':
        print("Closing Ticket Counter...")
        break
    else:
        print("Invalid choice! Please try again.")
```

**Linked List**:A Linked List is a linear data structure where elements (nodes) are connected using pointers.

* Data is stored in nodes

* Each node points to the next node

[Data | Next] → [Previous|Data | Next] → [Data | Next] → None

   10            20              30


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")


# Create object of LinkedList
ll = LinkedList()

# Insert elements
ll.insert_at_beginning(10)
ll.insert_at_beginning(5)
ll.insert_at_end(20)
ll.insert_at_end(30)

# Display list
ll.display()
```
